```python
#1. Get the numerical variables from our dataset.

numerical_data = data.select_dtypes(include=['float64', 'int64'])
```

```python
numerical_data
```

```python
#2. Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

import seaborn as sns
import matplotlib.pyplot as plt

#For normal distribution
for column in numerical_data.columns:
    sns.distplot(numerical_data[column])
    plt.title(f'Distribution of {column}')
    plt.show()
```

```python
#For exponential distribution
for column in numerical_data.columns:
    plt.hist(numerical_data[column], bins=20, density=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.show()
```

```python
#3. Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Log-transformed distribution plots
for column in numerical_data.columns:
    transformed_data = np.log(numerical_data[column])
    sns.distplot(transformed_data)
    plt.title(f'Log-Transformed Distribution of {column}')
    plt.show()
```
